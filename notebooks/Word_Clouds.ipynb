## Awesome Word Cloud Demo
This tutorial will demostrate how to make a word cloud in python and how to make pretty word clouds

```python
#/Users/stronglab2/Documents/NLP-and-Annotation/img
## Load Packages
## pip install [package here]
import numpy as np
from PIL import Image # This might be Pillow in Python 3.x
import matplotlib.pyplot as plt
from urllib2 import urlopen
from wordcloud import WordCloud, STOPWORDS
import nltk

%matplotlib inline
```

#### Download Text for Demo
For this demo we are using Herman Melville's Moby Dick

```python
## Download from URL
url = "http://www.gutenberg.org/files/2701/2701-0.txt"
response = urlopen(url)
raw = response.read().decode('utf8')
```

```python
## Download from Project Gutenberg 
nltk.download('gutenberg')
words = nltk.corpus.gutenberg.words('melville-moby_dick.txt')
```

```python
max(words, key=len) # find the longest word in the book!
```

#### Stopwords
Stopwords are conjunctions and common article are removed

```python
stopwords = set(STOPWORDS)
stopwords.add("said")
```

#### Generate Wordcloud
The first line initializes the parameters of the word cloud. The following lines generate and save the word cloud to a file location.

```python
wc = WordCloud(background_color="white", max_words=2000, stopwords=stopwords)

plt.imshow(wc.generate(raw))
plt.axis("off")
plt.show()
```

## Generate Whale Image Word Cloud

```python
## Load Image
whale_pic = np.array(Image.open('/Users/stronglab2/Documents/NLP-and-Annotation/img/whale.jpg'))
```

#### Added the mask parameter to the WordCloud Function

```python
wc2 = WordCloud(background_color="white", max_words=2000, mask=whale_pic, stopwords=stopwords)

plt.imshow(wc2.generate(raw))
plt.axis("off")
plt.show()
```
